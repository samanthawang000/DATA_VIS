## Background and Prior Work

The relationship between student populations and crime rates on university campuses has become an increasingly significant concern, particularly in California, where diverse and fluctuating student demographics create unique challenges. Universities like UC San Diego experience large variations in student presence throughout the year due to the academic calendar, seasonal enrollment patterns, and social dynamics. These changes may influence crime rates, making it essential to explore how fluctuations in student population correlate with campus crime trends. Understanding these patterns can help university administrators allocate safety resources more effectively and implement policies to enhance campus security.

Previous studies on campus crime trends suggest that crime rates tend to rise during periods of increased student activity. Research conducted by the National Center for Education Statistics has shown that crimes such as theft, burglary, and assault tend to be more frequent at the beginning of the academic year when campuses are more densely populated. Conversely, crime rates tend to decline during breaks when fewer students are present. However, these studies often focus on short-term crime fluctuations or aggregated data across multiple campuses, making it difficult to determine the specific impact of seasonal population shifts at individual institutions. Given that UCSD operates on a quarter system rather than the more common semester schedule, its student population dynamics may differ significantly from those of other universities.

UCSD enrollment data provides insight into these seasonal shifts. According to the UCSD Institutional Research office, total enrollment for the Fall 2023 quarter was 43,381 students, including 33,792 undergraduates and 9,589 graduate students. Enrollment patterns typically follow a cyclical trend, with significant drops during summer sessions when fewer courses are offered and a rise in student presence during the fall and winter quarters. The number of students physically present on campus may also vary based on housing policies, remote learning options, and commuting patterns, all of which influence the level of campus activity and potential crime risks.

Crime statistics from UCSD further illustrate the potential correlation between student population and crime rates. Data from the San Diego Association of Governments shows that in 2022, UCSD reported 537 property crimes and 20 crimes against society. In 2023, these numbers increased to 1,133 property crimes and 105 crimes against society, indicating a notable rise in criminal incidents. While these figures suggest a link between student density and crime levels, they do not account for external factors such as changes in law enforcement presence, campus security measures, or broader regional crime trends.

To better understand whether these crime fluctuations are unique to UCSD or part of a broader pattern, this study will compare campus crime rates with crime trends in the surrounding city of San Diego. By incorporating data from the San Diego Police Department’s crime reports, we will assess whether crime at UCSD increases in parallel with city-wide crime trends or if the university experiences distinct fluctuations related to student presence. If UCSD’s crime rate rises disproportionately compared to the city as a whole, it would indicate that factors such as student density and university policies play a more significant role in influencing campus crime. Conversely, if crime trends at UCSD mirror those seen across San Diego, it may suggest that broader socio-economic and policing factors drive the increase, rather than university-specific conditions.

Additionally, certain crimes may be more prevalent during specific times of the academic year. For example, at the start of a new quarter, campuses often experience increased foot traffic, social events, and late-night activities, potentially leading to a rise in crimes such as theft, alcohol-related offenses, and assault. During exam periods, stress and mental health challenges may contribute to different types of crime, such as vandalism or drug-related offenses. Understanding these patterns can help university administrators develop proactive safety strategies, such as adjusting security patrols, increasing awareness campaigns, and improving campus lighting in high-risk areas.

By systematically analyzing crime data from 2021 to 2023 alongside UCSD’s student enrollment trends and comparing these findings to city-wide crime trends, this study aims to determine whether fluctuations in student presence correspond to measurable changes in campus crime rates or if they align with broader regional crime patterns. The findings will provide valuable insights for campus safety planning and resource allocation, potentially guiding policies that enhance student security throughout the academic year.

## Timely Warnings and Community Alerts

This dataset includes all timely warnings and community alerts issued by the UCSD Police Department from 2019 to 2025, but only data from 2019 to 2023 is used for this analysis to focus on the most relevant period for detecting trends. The data was filtered to include only crime-related incidents such as burglary, sexual assault, robbery, vehicle theft, arson, weapons violations, aggravated assault, and indecent exposure. Non-crime-related incidents were excluded to ensure the analysis focuses solely on relevant trends.

The data was originally in a message format (police reports sent to UCSD-affiliated staff and students) and was converted into a CSV file for analysis. Categories were grouped by similarity to simplify the variables. The dataset is used to analyze how the frequency and type of crimes correlate with seasonal changes in student population.

```python
%matplotlib inline

# Loading and Preparing Data: The data was read from a CSV file and preprocessed to extract the year from the date and filter out non-crime-related incidents
# Additionally, a filter was applied to include only data from 2019 to 2023

DATA_FILE = "timely_warnings_2019_2025.csv"

def load_and_prepare_data():
    df = pd.read_csv(DATA_FILE)
    df["Year"] = pd.to_datetime(df["Date"], errors='coerce').dt.year
    df = df.dropna(subset=["Year"])
    df = df[(df["Year"] >= 2019) & (df["Year"] <= 2023)]  # Filter for 2019-2023
    df["Year"] = df["Year"].astype(int)
    return df

df = load_and_prepare_data()

#Filtering Non-Crime Reports: Specific keywords were used to exclude non-crime reports
NON_CRIME_KEYWORDS = [
    "Fire", "Suspect in Custody", "All Clear", "Campus Phone Outage", 
    "Security Report", "Trespassing", "Missing Juvenile", "Scams", 
    "Safety Report", "Catalytic Converter", "Unlawful Entry"
]

def filter_crime_reports(df):
    filtered_df = df[
        ~df["Description"].str.contains('|'.join(NON_CRIME_KEYWORDS), case=False, na=False)
    ].copy()
    return filtered_df

filtered_df = filter_crime_reports(df)

#Categorizing Crimes: Incidents were categorized based on the type of crime to simplify analysis

CRIME_CATEGORIES = {
    "Burglary": ["Burglary", "Burglary (REVISED)", "Attempted Burglary"],
    "Sexual Assault": ["Sexual Assault", "Sexual Battery", "Stalking and Sexual Assault – ARREST MADE"],
    "Robbery": ["Robbery", "Attempted Robbery"],
    "Vehicle Theft": ["Motor Vehicle Theft"],
    "Arson": ["Arson"],
    "Weapons Violation": ["Weapons Law Violation", "Brandishing of a Firearm"],
    "Aggravated Assault": ["Aggravated Assault"],
    "Indecent Exposure": ["Indecent Exposure"]
}

def categorize_crime(description):
    for category, keywords in CRIME_CATEGORIES.items():
        if any(keyword in description for keyword in keywords):
            return category
    return "Other"

def categorize_and_count(df):
    df["Crime Category"] = df["Description"].apply(categorize_crime)
    crimes_df = df[~df["Crime Category"].isin(["Avoid Area Alerts", "Other"])].copy()
    counts = crimes_df.groupby(["Year", "Crime Category"]).size().reset_index(name="Count")
    return counts

category_counts_by_year = categorize_and_count(filtered_df)

# This section includes visualizations and analysis based on the cleaned and categorized data to identify trends and patterns in crime rates relative to student population changes

def plot_crime_data_line(category_counts_by_year):
    plt.figure(figsize=(14, 8))
    for category in category_counts_by_year["Crime Category"].unique():
        subset = category_counts_by_year[category_counts_by_year["Crime Category"] == category]
        plt.plot(subset["Year"], subset["Count"], marker='o', label=category, alpha=0.7)
    
    plt.xticks(category_counts_by_year["Year"].unique())  # Force x-axis to show only integer years
    plt.xlabel("Year")
    plt.ylabel("Number of Incidents")
    plt.title("Timely Warnings Alerts (2019-2023)")
    plt.legend(title="Crime Category", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.show()

plot_crime_data_line(category_counts_by_year)
```

## Annual Security & Fire Safety Report

 This dataset includes crime data from UCSD between 2021 and 2023, focusing on property crimes (burglary, motor vehicle theft, and arson) and sexual offenses (rape, fondling, and aggravated assault). Non-relevant offenses (murder) were excluded for a more focused analysis.

The data was grouped into two categories: property crimes and sexual offenses. Bar charts were created to visualize the total occurrences of each crime type per year, highlighting trends between 2021 and 2023.

```python
# Loading and Preparing Data: Excludes non-relevant offenses and converts year column to integer format.
DATA_FILE = "UCSD_Crime_Statistics__2021-2023_.csv"
EXCLUDED_OFFENSES = ["Murder/Non-Negligent Manslaughter", "Incest", "Manslaughter by Negligence"]
PROPERTY_CRIMES = ["Burglary", "Motor Vehicle Theft", "Arson"]
VIOLENT_CRIMES = ["Rape", "Fondling", "Aggravated Assault"]

# Color mapping for different offenses
COLOR_MAP = {
    
    "Burglary": "#A7C7E7",               # Pastel Sky Blue
    "Motor Vehicle Theft": "#B2D7EE",    # Light Pastel Blue
    "Arson": "#779ECB",                  # Dark Pastel Blue
    
    
    "Rape": "#C8A2C8",                   # Lilac 
    "Fondling": "#D8BFD8",               # Thistle 
    "Aggravated Assault": "#E6E6FA"      # Lavender 
}


# Loading and cleaning data
def load_and_prepare_data(filepath):
    df = pd.read_csv(filepath)
    df = df[~df['Offense'].isin(EXCLUDED_OFFENSES)]  # Exclude irrelevant offenses
    df['Year'] = df['Year'].astype(int)  # Convert year to integer
    return df

# Filter data based on offense categories
def filter_data_by_offenses(df, offenses):
    return df[df['Offense'].isin(offenses)]

# Plot bar charts for crime data
def plot_crime_data_bar(df_subset, offenses, title):
    plt.figure(figsize=(12, 8))
    width = 0.2  # Width of each bar
    years = sorted(df_subset['Year'].unique())
    x = range(len(years))
    
    for idx, offense in enumerate(offenses):
        subset = df_subset[df_subset['Offense'] == offense]
        plt.bar(
            [i + idx * width for i in x], subset['Total'],
            width=width, label=offense,
            color=COLOR_MAP.get(offense, "gray"), alpha=0.8
        )
    
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Total Crimes", fontsize=12)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xticks([i + width for i in x], years)
    plt.legend(title="Offense", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Main section to execute the plotting
if __name__ == "__main__":
    df = load_and_prepare_data(DATA_FILE)
    
    df_property = filter_data_by_offenses(df, PROPERTY_CRIMES)
    df_sexual = filter_data_by_offenses(df, VIOLENT_CRIMES)
    
    plot_crime_data_bar(df_property, PROPERTY_CRIMES, "Property Crimes at UCSD (2021-2023)")
    plot_crime_data_bar(df_sexual, VIOLENT_CRIMES, "Sexual Offenses and Assaults at UCSD (2021-2023)")
```

## San Diego Police Department Crime Data Explorer

This dataset contains crime incidents reported in La Jolla, San Diego, between 2021 and 2023, covering offenses such as theft, assault, burglary, fraud, drug violations, and property crimes. To provide a clearer understanding of overall crime trends, similar offenses were grouped into broader categories, allowing for a more structured analysis of crime patterns.

Each category represents a set of related offenses, including Larceny/Theft, Assault & Violence, Property Crimes, Motor Vehicle Crimes, Fraud & Financial Crimes, Drug & Alcohol Violations, and Public Order Offenses. The total number of reported incidents within each category was then calculated to highlight the most prevalent crime types during the study period.

A bar chart was chosen to visualize these trends, allowing for a direct comparison of crime frequency across different categories for both 2021 and 2023. The color scheme was optimized for readability, making it easier to identify which crimes were most common and how they compare over time.

```python
file_path = "Crime_Data_2021_2023.csv"
df = pd.read_csv(file_path)

# Define crime categories similar to your reference
CRIME_CATEGORIES = {
    "Larceny/Theft": ["All Other Larceny", "Theft From Motor Vehicle", "Shoplifting", "Pocket-picking", 
                      "Purse-snatching", "Theft From Building", "Stolen Property Offenses",
                      "Theft of Motor Vehicle Parts/Accessories"],

    "Burglary/Robbery": ["Burglary/Breaking & Entering", "Robbery"],

    "Motor Vehicle Crimes": ["Motor Vehicle Theft"],

    "Assault & Violence": ["Simple Assault", "Aggravated Assault", "Forcible Fondling", "Forcible Rape", 
                           "Forcible Sodomy", "Sexual Assault", "Indecent Exposure"],

    "Property Crimes": ["Destruction/Damage/Vandalism of Property", "Arson", "Trespass of Real Property"],

    "Fraud & Financial Crimes": ["Impersonation", "Credit Card/Automatic Teller Fraud", 
                                 "Counterfeiting/Forgery", "False Pretenses/Swindle/Confidence Game", 
                                 "Embezzlement", "Identity Theft"],

    "Drug & Alcohol Violations": ["Drunkenness", "Drug/Narcotic Violations", "Drug Equipment Violations",
                                  "Liquor Law Violations", "Driving Under the Influence"],

    "Weapons & Violent Crimes": ["Weapons Violation", "Kidnapping/Abduction", "Murder and Nonnegligent Manslaughter",
                                 "Extortion/Blackmail"],

    "Public Order Offenses": ["Disorderly Conduct", "Curfew/Loitering/Vagrancy Violations",
                              "Prostitution", "Pornography/Obscene Material"],

    "Other Offenses": ["All Other Offenses", "Intimidation", "Peeping Tom", "Statutory Rape", "Incest"]
}

# Function to categorize crimes
def categorize_crime(crime_type):
    for category, offenses in CRIME_CATEGORIES.items():
        if crime_type in offenses:
            return category
    return "Other"

# Filter for 2021 and 2023 crimes
df_filtered = df[df["Year"].isin([2021, 2023])].copy()
df_filtered["Crime Category"] = df_filtered["Crime Type"].apply(categorize_crime)

# Separate data for 2021 and 2023
df_2021 = df_filtered[df_filtered["Year"] == 2021].groupby("Crime Category")["Total Offenses"].sum().reset_index()
df_2023 = df_filtered[df_filtered["Year"] == 2023].groupby("Crime Category")["Total Offenses"].sum().reset_index()

# Merge both years into a single DataFrame for comparison
df_combined = df_2021.merge(df_2023, on="Crime Category", how="outer", suffixes=("_2021", "_2023")).fillna(0)

# Display the categorized data
print(df_combined)  # Print the dataframe for review

# Plot comparison of crime categories between 2021 and 2023
plt.figure(figsize=(14, 7))
df_combined.set_index("Crime Category").plot(kind="bar", figsize=(14, 7), alpha=0.85, color=["#1f77b4", "#ff474c"])
plt.xlabel("Crime Category")
plt.ylabel("Total Reported Crimes")
plt.title("Comparison of Crime Incidents in La Jolla 2021 vs 2023")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.legend(["2021", "2023"])
plt.show()
```

This following code was designed to analyze the relationship between UCSD's student population and crime rates at both UCSD and La Jolla for 2021 and 2023. First, the student population dataset was cleaned by extracting the year from the "Quarter" column and calculating the average on-campus student population per year. This allowed us to observe trends at a yearly level rather than by academic term. The UCSD crime dataset was then processed by summing total reported crimes for each year, creating an annual total for better comparison. Similarly, the La Jolla crime dataset was filtered to only include 2021 and 2023, ensuring a direct comparison with UCSD data.

The three datasets were merged by Year, aligning student population, UCSD crime, and La Jolla crime data for side-by-side analysis. A bar chart was used to clearly visualize these trends: on-campus student population (blue, left y-axis), UCSD crime rates (red, right y-axis), and La Jolla crime rates (green, right y-axis, different scale). This approach allowed us to see that UCSD crime increased as student population grew, while La Jolla’s crime trends remained unrelated to these changes. The results reinforce that campus crime is more influenced by student density and university-specific factors rather than broader regional crime trends.

```python
student_population_file = "UCSD_Student_Population_by_Quarter.csv"
df_students = pd.read_csv(student_population_file)

df_students = df_students.rename(columns={
    "On-Campus Students": "On_Campus_Students",
    "Total Enrollment": "Total_Enrollment"
})

df_students["Year"] = df_students["Quarter"].str.extract(r'(\d{4})')[0].astype(int)

df_on_campus = df_students.groupby("Year")["On_Campus_Students"].mean().reset_index()

crime_stats_file = "UCSD_Crime_Statistics__2021-2023_.csv"
df_crime = pd.read_csv(crime_stats_file)

df_crime["Year"] = df_crime["Year"].astype(int)
df_crime_counts = df_crime.groupby("Year")["Total"].sum().reset_index()
df_crime_counts = df_crime_counts[df_crime_counts["Year"].between(2021, 2023)]

lajolla_crime_file = "Crime_Data_2021_2023.csv"
df_lajolla_crime = pd.read_csv(lajolla_crime_file)

df_lajolla_crime = df_lajolla_crime[df_lajolla_crime["Year"].isin([2021, 2023])]
df_lajolla_crime_summary = df_lajolla_crime.groupby("Year")["Total Offenses"].sum().reset_index()

df_merged = pd.merge(df_on_campus, df_crime_counts, on="Year", how="inner")
df_merged = pd.merge(df_merged, df_lajolla_crime_summary, on="Year", how="inner")

bar_width = 0.3
years = np.arange(len(df_merged["Year"]))

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(years - bar_width, df_merged["On_Campus_Students"], width=bar_width, label="On-Campus Students", color="blue", alpha=0.7)
ax1.set_ylabel("On-Campus Student Population", color="blue")
ax1.set_xlabel("Year")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()
ax2.bar(years, df_merged["Total"], width=bar_width, label="Total UCSD Crimes", color="red", alpha=0.7)
ax2.set_ylabel("Total Crime Reports", color="red")
ax2.tick_params(axis='y', labelcolor="red")

ax3 = ax1.twinx()
ax3.spines["right"].set_position(("outward", 60))
ax3.bar(years + bar_width, df_merged["Total Offenses"], width=bar_width, label="La Jolla Crimes", color="green", alpha=0.7)
ax3.set_ylabel("Total La Jolla Crimes", color="green")
ax3.tick_params(axis='y', labelcolor="green")

ax1.set_xticks(years)
ax1.set_xticklabels(df_merged["Year"])
ax1.set_title("On-Campus Student Population vs. Crime Reports Per Year (UCSD & La Jolla)")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
ax3.legend(loc="lower right")
plt.grid(axis='y', linestyle="--", alpha=0.5)
plt.show()
```

## Seperating Crime Rate Data

# Total Crimes at UCSD
Here, we have graphically represented the total number of crimes reported in both the Timely Warnings and the Annual Report datasets from 2021 to 2023. By doing so, we are able to get a visualization of how total crime has increased and whether or not there are any correlations between the two. The Annual Report shows a significant and steady increase in crime over the years, rising from 181 in 2021 to 589 in 2023. In contrast, crimes reported through Timely Warnings are much lower but still show a gradual increase from 7 in 2021 to 20 in 2023.

```python
TIMELY_WARNINGS_FILE = "timely_warnings_2019_2025.csv"
CRIME_STATS_FILE = "UCSD_Crime_Statistics__2021-2023_.csv"

# Exclude non-crime reports for Timely Warnings
NON_CRIME_KEYWORDS = [
    "Fire", "Suspect in Custody", "All Clear", "Campus Phone Outage", 
    "Security Report", "Trespassing", "Missing Juvenile", "Scams", 
    "Safety Report", "Catalytic Converter", "Unlawful Entry"
]

# Crime categories for filtering
CRIME_CATEGORIES = [
    "Burglary", "Motor Vehicle Theft", "Arson", 
    "Rape", "Fondling", "Aggravated Assault"
]

def load_and_prepare_timely_warnings(filepath):
    df = pd.read_csv(filepath)
    df["Year"] = pd.to_datetime(df["Date"], errors='coerce').dt.year
    df = df.dropna(subset=["Year"])
    df = df[(df["Year"] >= 2021) & (df["Year"] <= 2023)]  # Filter for 2021-2023
    df = df[~df["Description"].str.contains('|'.join(NON_CRIME_KEYWORDS), case=False, na=False)]
    df["Year"] = df["Year"].astype(int)
    df['Total'] = 1  # Each entry is a single incident
    total_crimes_per_year = df.groupby('Year').size().reset_index(name='Total')
    total_crimes_per_year['Source'] = 'Timely Warnings'
    return total_crimes_per_year

def load_and_prepare_crime_stats(filepath):
    df = pd.read_csv(filepath)
    df = df[df['Offense'].isin(CRIME_CATEGORIES)]
    df = df[(df['Year'] >= 2021) & (df['Year'] <= 2023)]  # Filter for 2021-2023
    df['Year'] = df['Year'].astype(int)
    total_crimes_per_year = df.groupby('Year')['Total'].sum().reset_index()
    total_crimes_per_year['Source'] = 'Annual Report'
    return total_crimes_per_year

def merge_datasets(timely_warnings_df, crime_stats_df):
    merged_df = pd.concat([timely_warnings_df, crime_stats_df], ignore_index=True)
    return merged_df

# Plot combined data
def plot_combined_data(merged_df):
    plt.figure(figsize=(10, 6))

    for source, color, linestyle in zip(
        ["Timely Warnings", "Annual Report"], 
        ['#6495ED', '#FFAB76'], 
        ['-', '--']
    ):
        subset = merged_df[merged_df['Source'] == source]
        plt.plot(
            subset['Year'], subset['Total'], 
            marker='o', linestyle=linestyle, linewidth=2, label=source, color=color, alpha=0.8
        )
        plt.fill_between(subset['Year'], subset['Total'], color=color, alpha=0.2)

        # Add data labels for clarity
        for x, y in zip(subset['Year'], subset['Total']):
            plt.text(x, y + 5, str(y), ha='center', fontsize=10, color=color)

    plt.xticks([2021, 2022, 2023], fontsize=12)  
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Total Crimes", fontsize=12)
    plt.title("Total Crimes: Timely Warnings vs. Annual Report (2021-2023)", fontsize=14, fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(title="Source")
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    timely_warnings_df = load_and_prepare_timely_warnings(TIMELY_WARNINGS_FILE)
    crime_stats_df = load_and_prepare_crime_stats(CRIME_STATS_FILE)
    merged_df = merge_datasets(timely_warnings_df, crime_stats_df)
    plot_combined_data(merged_df)
```

## 2023 Spike in Crime Rate and On-Campus Population

This shows the correlation between on-campus student population and total crime reports from the UCSD Annual Report for the years 2021 to 2023 using a dual-axis bar chart. It first loads and processes the student population dataset, renaming columns for consistency and extracting the year from the "Quarter" column. The data is then grouped by year to compute the average on-campus student population. By loading the UCSD crime statistics dataset and filtering only for 2021 to 2023, we are able to sum the total reported crimes per year. The two datasets are merged on the "Year" column, ensuring alignment for comparison. The visualization is created using side-by-side bars, where the blue bars represent on-campus student population (left y-axis) and the red bars represent total crime reports (right y-axis). The dual-axis format give a direct visual correlation between student population size and crime rates. By displaying this data, it highlighs the notable increase in crime alongside on-campus student growth in 2023 compared to 2021.

```python
student_population_file = "UCSD_Student_Population_by_Quarter.csv"
df_students = pd.read_csv(student_population_file)

df_students = df_students.rename(columns={
    "On-Campus Students": "On_Campus_Students",
    "Total Enrollment": "Total_Enrollment"
})

df_students["Year"] = df_students["Quarter"].str.extract(r'(\d{4})')[0].astype(int)

df_on_campus = df_students.groupby("Year")["On_Campus_Students"].mean().reset_index()

df_crime["Year"] = df_crime["Year"].astype(int)
df_crime_counts = df_crime.groupby("Year")["Total"].sum().reset_index()
df_crime_counts = df_crime_counts[df_crime_counts["Year"].between(2021, 2023)]

df_merged = pd.merge(df_on_campus, df_crime_counts, on="Year", how="inner")

bar_width = 0.4
years = np.arange(len(df_merged["Year"]))

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(years - bar_width/2, df_merged["On_Campus_Students"], width=bar_width, label="On-Campus Students", color="blue", alpha=0.7)
ax1.set_ylabel("On-Campus Student Population", color="blue")
ax1.set_xlabel("Year")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()
ax2.bar(years + bar_width/2, df_merged["Total"], width=bar_width, label="Total Crime Reports (Annual Report)", color="red", alpha=0.7)
ax2.set_ylabel("Total Crime Reports", color="red")
ax2.tick_params(axis='y', labelcolor="red")

ax1.set_xticks(years)
ax1.set_xticklabels(df_merged["Year"])
ax1.set_title("On-Campus Student Population vs. Total Crime Reports Per Year (Annual Report)")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.grid(axis='y', linestyle="--", alpha=0.5)
plt.show()
```

## Student Population and Crime Statistics Correlation Heatmap

This correlation heatmap provides a visual representation of the relationships between UCSD student population statistics and crime occurrences from 2021 to 2023. By analyzing the correlations between variables such as total student enrollment, on-campus student housing, and different crime locations (on-campus, non-campus, and public property), this heatmap helps identify potential trends in crime rates relative to student population changes. A positive correlation between total students and public property crimes, for example, might suggest that a larger student presence contributes to increased incidents in public areas, whereas a weak correlation with on-campus crimes could indicate effective campus security measures. This analysis is valuable for university administrators and policymakers, providing insights that can help inform crime prevention strategies, resource allocation, and campus safety improvements.

```python
crime_data_path = "UCSD_Crime_Statistics__2021-2023_.csv"
student_population_path = "UCSD Student Population.csv"

def load_and_prepare_crime_data(filepath):
    df = pd.read_csv(filepath)
    df['Year'] = df['Year'].astype(str) 
    return df

def load_and_prepare_population_data(filepath):
    df = pd.read_csv(filepath)
    df['Year'] = df['Year'].astype(str).str[:4]  
    return df

crime_data = load_and_prepare_crime_data(crime_data_path)
student_population = load_and_prepare_population_data(student_population_path)
merged_data = pd.merge(crime_data, student_population, on='Year', how='inner')

columns_to_exclude = ["Bachelor's Degrees Awarded", "Master's Degrees Awarded", 
                      "Doctoral Degrees Awarded (Research/Scholarship)", 
                      "Doctoral Degrees Awarded (Professional Practice)"]
correlation_data = merged_data.drop(columns=columns_to_exclude, errors='ignore').select_dtypes(include=['number'])

correlation_matrix = correlation_data.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap: UCSD Student Population vs Crime Statistics")
plt.show()
```

Some key observations on Student Population vs. Crime:
- Total students and public property crimes have a moderate positive correlation (0.65), which indicates that as the total student population increases, public property crimes also increase.
- Total students and on-campus crimes have a weak positive correlation (0.17), which indicates that a larger student population may contribute to slightly higher on-campus crime
- Total graduate students and total crimes have a weak correlation (0.19), which suggests that graduate student numbers slightly impact crime

Some key observations on Crime Locations:
- Public property crimes and on-campus total crimes have a moderate positive correlation (0.62), which suggests that when crimes occur on public property, they are often accompanied by an increase in on-campus crimes
- Public property crimes and off-campus crimes have a moderate correlation (0.47), which suggests that crimes occuring in non-campus areas correlate moderately with crimes on public property

Overall, the more students present on campus, the more likely public property crimes are committed. On-campus crime rates are only weakly correlated with student population, and public property and non-campus crimes are moderately related.
